```python
mapper = {"Sellin Cost":"Selling Cost"}
products.rename(columns=mapper,inplace=True)
```

```python
#Total buying cost and selling cost by each department
dept_bs = products.groupby(["Product Category"])[["Product Cost","Selling Cost"]].sum().sort_values("Selling Cost",ascending=False)
dept_bs.plot(kind="bar")
plt.grid(False)
plt.title("Buy & Sell Cost")
plt.xlabel("Product Department",labelpad=15)
plt.ylabel("Cost",labelpad=15)
plt.show()
```

```python
#Total Number Of Products Per Department
dept_products = products.groupby(["Product Category"])["Product Name"].count()
dept_products.sort_values(ascending=True).plot(kind="barh",grid=False)
plt.xlabel("Product Count",labelpad=15)
plt.ylabel("Product Department",labelpad=15)
plt.show()
```

```python
# Product Buying Cost,Selling Cost and Profit Margin By Each Departments
df_bsm = (products.groupby(["Product Category"])[["Product Cost","Selling Cost"]].
          sum().
          assign(margin = lambda x: x['Selling Cost'] - x['Product Cost']))
```

```python
df_bsm.plot(kind="area",figsize=(15,5),grid=False)
plt.title("Product Profit Margin")
plt.show()
```

```python
#Revenue and Profits Per Year
rev_yr = purchases.merge(orders,on="Bill ID").merge(products,on="Product ID")
```

```python
rev_yr["Revenue"] = rev_yr["Quantity"] * rev_yr["Selling Cost"]
```

```python
rev_yr["Profits"] = rev_yr["Revenue"] - rev_yr["Product Cost"]
```

```python
rev_yr
```

```python
#Most Profitable Products Departments
rcp_dept = (rev_yr.groupby(["Product Category"])
            [["Product Cost","Revenue","Profits"]].sum().sort_values("Revenue",ascending=False))
rcp_dept
```

```python
rcp_dept.plot(kind="bar",grid=False)
plt.title("Revenue,Cost&Profits By Dept")
plt.xlabel("Departments",labelpad=15)
plt.ylabel("Values in Cr",labelpad=15)
```

```python
#Most Profitable Products
rcp_product = (rev_yr.groupby(["Product Name"])
               [["Revenue","Product Cost","Profits"]].sum().sort_values("Revenue",ascending=False))
```

```python
rcp_product.plot(kind="bar",figsize=(15,5),grid=False)
plt.title("Revenue,cost&Profits By Products")
```

```python
#Revenue,Cost and Profits By Year
rcp_year = rev_yr.groupby(["Year"])[["Product Cost","Revenue","Profits"]].sum()
rcp_year
```

```python
rcp_year.plot(kind="barh",grid=False,figsize=(10,5))
plt.title("Revenue,Cost&Profits By Year")
plt.xlabel("values in Cr",labelpad=15)
plt.show()
```
